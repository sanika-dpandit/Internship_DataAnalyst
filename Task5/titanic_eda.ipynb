```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace path with your file location if needed)
df = pd.read_csv("titanic.csv")

# For better plot style
sns.set(style="whitegrid")
```

```python
# First few rows
df.head()
```

```python
# Dataset info
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Value counts for target variable
df['Survived'].value_counts()
```

```python
sns.histplot(df['Age'], kde=True, bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

- The Age distribution is slightly right-skewed.
- Most passengers are between 2040 years old.

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

- First-class passengers tend to be older on average.
- Third-class passengers show a wider spread in age distribution.

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare with Survival")
plt.show()
```

- Higher fares seem to be associated with higher survival.
- Younger passengers are spread across all fare ranges.

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived')
plt.show()
```

- There is a visible separation in Fare for survivors vs non-survivors.

```python
numeric_df = df.select_dtypes(include=['number'])  # Only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

- Fare has a moderate positive correlation with Survival.
- Pclass has a negative correlation with Survival.

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Gender")
plt.show()
```

- Females had a much higher survival rate compared to males.

### Summary of Findings:
- Higher Fare and 1st Class travel were linked to better survival chances.
- Women and younger passengers had higher survival rates.
- Age distribution is slightly skewed towards younger passengers.
- Pclass and Fare are important predictors of survival.
- There are some missing values in Age and Cabin columns.
